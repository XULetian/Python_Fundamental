## 5

```python
house = np.genfromtxt('housing.csv', delimiter=',',
                      dtype=['<f8','<f8','<i4','<i4','<i4','<i4','<i4','<f8','<i4','U15'])
```

```python
house_attr = ['longitude',
              'latitude',
              'housing_median_age',
              'total_rooms',
              'total_bedrooms',
              'population',
              'households',
              'median_income',
              'median_house_value',
              'ocean_proximity']
```

```python
house = house[1:]
house.size
```

## 6

```python
for i in range(0, 10, 1):
    print("Data type for", house_attr[i], "attribute is: ", (house[0][i]).dtype) 
```

## 7 

```python
newhouse = []
for row in house:
    newhouse_2 = []
    for i in row:
        newhouse_2.append(i)
    newhouse.append(newhouse_2)
    
house = np.array(newhouse)
house.shape
```

## 8

```python
import matplotlib.pyplot as plt
```

```python
population = []
index = house_attr.index('population')
for row in house:
    population.append(int(row[index]))
```

```python
plt.hist(population, bins='auto')  # arguments are passed to np.histogram
plt.xlabel('Population')
plt.ylabel('Count')
plt.title("Histogram of Population")
plt.show()
```

## 9

```python
ocean = []
index = house_attr.index('ocean_proximity')
for row in house:
    ocean.append(row[index])
```

```python
ocean_kind = set()
for row in ocean:
    ocean_kind.add(row)
```

```python
ocean_kind
```

```python
ocean_number = [ocean.count('<1H OCEAN'), 
                ocean.count('INLAND'),
                ocean.count('ISLAND'),
                ocean.count('NEAR BAY'),
                ocean.count('NEAR OCEAN')]
```

```python
plt.bar(list(ocean_kind), ocean_number)
plt.xlabel('Ocean Proximity')
plt.ylabel('Count')
plt.title("Histogram of Ocean Proximity")
plt.show()
```

## 10

```python
age = []
index = house_attr.index('housing_median_age')
for row in house:
    age.append(int(row[index]))
```

```python
price = []
index = house_attr.index('median_house_value')
for row in house:
    price.append(int(row[index]))
```

```python
plt.figure(figsize=(10,10)) 
plt.scatter(age, population, c = price)
plt.xlabel('Housing Median Age')
plt.ylabel('Population')
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

### It's hard to identify any hidden relations between price, population and age from the plot above. The first conclusion that we can make is that the housing meidan age younger, the number of cases is bigger. Then we have to plot the three attribute with their location to find out some potential hidden relations.

```python
long = []
index = house_attr.index('longitude')
for row in house:
    long.append(float(row[index]))

lat = []
index = house_attr.index('latitude')
for row in house:
    lat.append(float(row[index]))
```

```python
plt.figure(figsize=(10,10)) 
plt.scatter(long, lat, c = price)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10)) 
plt.scatter(long, lat, c = age)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10)) 
plt.scatter(long, lat, c = population)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Scatter Plot")
plt.colorbar()
plt.show()
```

### From three plot above, we can find out that most expensive housing price in California area is located around two major city, San Fransico, and Los Angeles. Also, these two cities have the most population and relatively older housing median age. That we can assume that the house value has the positive linear relationship with housing age and population.
